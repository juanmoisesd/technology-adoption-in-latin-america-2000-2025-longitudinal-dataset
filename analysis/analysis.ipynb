Longitudinal dataset (2000-2025) on digital technology adoption indicators across 21 Latin American and Iberian countries. Variables: internet penetration, mobile subscriptions, smartphone adoption, social media usage, e-commerce, digital payments, AI tool usage, digital skills index, and technology anxiety prevalence. Sources: ITU, World Bank, GSMA, Statista.

**Keywords:** technology adoption, digital transformation, Latin America, internet penetration, smartphone, AI adoption, longitudinal dataset


```python
# Install dependencies if needed
# !pip install pandas matplotlib seaborn numpy

import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import warnings
warnings.filterwarnings('ignore')

plt.rcParams['figure.dpi'] = 120
print('Libraries loaded.')
```

## 1. Load Dataset


```python
# Download from: https://doi.org/10.5281/zenodo.18984814
# df = pd.read_csv('dataset.csv')

# Synthetic example data
np.random.seed(42)
years = list(range(2000, 2024))
keywords = ['technology adoption', 'digital transformation', 'Latin America', 'internet penetration']
df = pd.DataFrame({'year': years,
    **{k.replace(' ','_')[:12]: np.random.normal(50,15,len(years)) for k in keywords}})

print(f'Dataset: Technology Adoption in Latin America (2000-2025): ')
print(f'DOI: 10.5281/zenodo.18984814')
print(f'Shape: {df.shape}')
df.head()
```

## 2. Summary Statistics


```python
df.describe().round(2)
```

## 3. Temporal Trends


```python
fig, ax = plt.subplots(figsize=(10,5))
for col in df.columns[1:]:
    ax.plot(df['year'], df[col], marker='o', markersize=3, label=col)
ax.set_title('Technology Adoption in Latin America (2000-2025): Longi')
ax.set_xlabel('Year')
ax.set_ylabel('Value')
ax.legend(fontsize=8)
ax.grid(alpha=0.3)
plt.tight_layout()
plt.show()
```
